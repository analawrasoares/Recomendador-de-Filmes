```python
#IMPORTA DATASET
import pandas as pd
movie_data = pd.read_csv('movies.csv', low_memory=False) 
movie_data.head()
```

```python
#MOSTRA OS PRIMEIROS 5 OBJETOS DO BANCO DE DADOS
movie_data['overview'].head(5)
```

```python
#IMPORTA O sci-kit modulo
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vector = TfidfVectorizer(stop_words='english')
#remove palavras como "the", 'A', etc
movie_data['overview'] = movie_data['overview'].fillna('')
#muda espaços vazios por uma string vazia para evitar erros no codigo
tfidf_matrix = tfidf_vector.fit_transform(movie_data['overview'])
#constroe matriz e transforma os dados
```

```python
#importa o linar_kernel para criar um matriz simetrica
from sklearn.metrics.pairwise import linear_kernel
sim_matrix = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
#cria um indice reverso para mostrar os filmes e também remove os filmes  duplicados
indices = pd.Series(movie_data.index,
index=movie_data['title']).drop_duplicates()
indices[:10]
```

```python
#função de recomendação 
def recomendar(title, sim_scores=sim_matrix):
    idx = indices[title]
    sim_scores = list(enumerate(sim_matrix[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]
    movie_indices = [i[0] for i in sim_scores]
    return movie_data['title'].iloc[movie_indices]
```

```python
#chama função com o nome do filme que você receber recomendações
recomendar('Toy Story')
```
